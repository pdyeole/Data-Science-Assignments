Q3) What errors are defined in the ArithmeticError class? Explain any two with an example

Arithmetic Error Class: The base class for those built in exceptions that are raised for various arithmetic errors: Overflow error, ZeroDivisionError, FloatingPointError
1) ZeroDivisionError: This error is raised when you try to divide by zero
2) Overflow Error: The error is raised when the result of an arithmetic operation is too large to be represented by the data type
3) FloatingPointError: This error is raised when an arithmetic operation involving floating-point numbers result in an invalid result

```python
import math
try:
    math.exp(1000)
except OverflowError as e:
    print(e)
```

Above program shows the case of overfloe error


```python
try:
    a= 10
    a/=0
except ZeroDivisionError as e:
    print(e)
```

Q4) Why lookup class is used? Explain with an example KeyError and IndexError

LookupError Exception is the base class for errors raised when something is not found.  The base class for the exceptions that are raised when a key or index used on mapping or a sequence is invalid: IndexError, Keyerror

```python
try:
    l=[1,2,3]
    print(l[3])
except IndexError as e:
    print(e)
```

```python
try:
    d= {"name":"pankaj","mobile":17139211}
    print(d['key'])
except KeyError as e:
    print(e)
```

Q5) Explain ImportError. What is ModuleNotFoundError?

Import Error is a type of error which occurs when python is unable to import a module. This error can occur due to variety of reasons such as missing or incorrect module name, or missing dependency
ModuleNotFound Error is a python error that occurs when you try to import a module that is not installed or not found in the current python path. This error occurs when you are working with external packages or libraries that are not included in the default python installation
